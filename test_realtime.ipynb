```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_json('/Users/ashwin.gangadhar/demo/maap-chatbot-builder/builder/partnerproduct/data/test.jsonl', lines=True)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
from confluent_kafka import Producer
import json
from time import sleep

# Kafka configuration
conf = {
    'bootstrap.servers': 'localhost:9092'
}

# Create Producer instance
producer = Producer(conf)

# Function to publish messages to Kafka
def publish_to_kafka(topic, message):
    producer.produce(topic, value=message)
    producer.flush()

# Publish each row to Kafka topic
for index, row in df.iterrows():
    message = row.to_json()
    sleep(2)
    publish_to_kafka('sink_mongo_topic', message)
```
